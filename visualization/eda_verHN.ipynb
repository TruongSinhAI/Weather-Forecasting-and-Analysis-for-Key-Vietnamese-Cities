```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
path = 'E:\HaoDevAI\DAP391m_Project\datasets\Hanoi.csv'
df = pd.read_csv(path, parse_dates=['datetime'])
print(path,' loaded successfully. Shape:', df.shape)
```

```python
cutoff_date = pd.Timestamp("2010-01-01")
df = df[df['datetime'] >= cutoff_date]
df.shape
```

```python
drop_cols = ['name','snow', 'snowdepth','sealevelpressure','icon','stations','sunrise','sunset']
df.drop(drop_cols, axis=1, inplace=True)
print("Irrelevant columns droped!")
```

```python
df.nunique()
```

```python
check_cols = ['precipprob','preciptype','uvindex','severerisk','conditions']
for col in check_cols:
    print(f"Unique values in '{col}':")
    print(df[col].unique())
    print(f"Counts values in '{col}':")
    print(df[col].count(),'/5496')
    print("-" * 30)
```

```python
#fill missing value
df['preciptype'] = df['preciptype'].apply(lambda x: True if x == 'rain' else False) #chuyen ve boolean rain=1 norain=0
df['severerisk']= df['severerisk'].fillna(value=0).infer_objects(copy=False) #fill NaNs bang 0 (no risk)
df['severerisk'] = df['severerisk'].apply(pd.to_numeric, errors='coerce')
    #fill windust with avg seasonal value
df['month_day'] = df['datetime'].dt.strftime('%m-%d')
df['windgust'] = df.groupby('month_day')['windgust'].transform(lambda x: x.fillna(x.mean()))
df.drop(columns=['month_day'], inplace=True)
#add new columns
df['temp_range'] = df['tempmax'] - df['tempmin']
df['month'] = df['datetime'].dt.month

#onehot encoding string columns
df_onehot = pd.get_dummies(df, columns=['conditions'])
df_onehot.info()
```

```python
plt.figure(figsize = (20,20))
temp = df_onehot
temp = temp.drop(columns=['preciptype','description'])
sns.heatmap(temp.corr(), annot=True, fmt='.2f', cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='datetime', y='temp', data=df)
plt.title('Temperature Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.show()
```

```python
cols = ['temp', 'humidity', 'windspeed', 'precip']
plt.figure(figsize=(12,8))
for i, col in enumerate(cols, 1):
    plt.subplot(2,2,i)
    sns.histplot(df[col], kde=True)
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.boxplot(x='month', y='temp', data=df)
plt.title('Monthly Temperature Distribution')
plt.xlabel('Month')
plt.ylabel('Temperature (°C)')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(y='conditions', data=df, order=df['conditions'].value_counts().index)
plt.title('Frequency of Weather Conditions')
plt.xlabel('Count')
plt.ylabel('Condition')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(x='datetime', y='winddir', data=df, hue='windspeed')
plt.title('Wind Direction Over Time')
plt.xlabel('Date')
plt.ylabel('Wind Direction (degrees)')
plt.show()
```
